```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
PATH_MUSHROOM_RESULT = "./mlruns/13/dfeaa754f89e45e09f13125a631941b4/artifacts/all_results.csv"
PATH_BANKNOTE_RESULT = "./mlruns/13/09f66b5c8ba247fa81a0430b5adc54f4/artifacts/all_results.csv"
PATH_DEFAULT_RESULT = "./mlruns/13/c1a42e2446ee46aba013033a5f6cf643/artifacts/all_results.csv"
PATH_CONCRETE_RESULT = "./mlruns/13/edb77488e1d84e908d8faf62d0fb10a4/artifacts/all_results.csv"
PATH_QSAR_RESULT = "./mlruns/13/31ebb4d78fcb40f087004453faf3317c/artifacts/all_results.csv"
PATH_BIKE_RESULT = "./mlruns/13/bde23952e5fa45ad9f571c8c5d185bd0/artifacts/all_results.csv"
PATH_TRAFFIC_RESULT = "./mlruns/13/eb65c27c1b0c4ceb9644fcc34e02b26a/artifacts/all_results.csv"
```

```python
CHANGE_NAMES = {"two-stage TrAdaBoost" : "two-stage", "Trbagg" : "TrBagg", "DART for DA" : "TrDART"}

PALETTE ={
    "target GBDT":"C0", 
    "all GBDT":"C1",
    "TrAdaBoost":"C2",
    "two-stage": "C2",
    "TrBagg": "C4",
    "only Dropout": "C5",
    "target DART": "C6",
    "TrDART": "C3"
}
```

```python
key_cols = ["n_targets","target_domain"]
```

# 各データセットごとに分析

## mushroom

```python
mean_mushroom = pd.read_csv(PATH_MUSHROOM_RESULT).groupby(key_cols).mean()\
                    .stack().reset_index()
mean_mushroom.rename(columns={"level_2":"method", 0:"auc"}, inplace=True)
for key, val in CHANGE_NAMES.items():
    mean_mushroom["method"].where(mean_mushroom["method"]!=key, val, inplace=True)
sns.relplot(x="n_targets", y="auc", hue="method", col="target_domain", kind="line", 
            palette=PALETTE, data=mean_mushroom)
```
